```python
import numpy as np

# Read the data from CSV file and save into an array of float numbers
data = np.loadtxt('M3 [Exercise] example_data.csv', delimiter=',', skiprows=1)
print(data)
```

```python
import matplotlib.pyplot as plt
import math

size_of_interval = math.sqrt(len(data))
size_of_interval = math.ceil(size_of_interval)
size_of_interval = int(size_of_interval)

# Draw a histogram for the data
plt.hist(data, bins=size_of_interval, edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
```

## Q-Q Plot

**Algorithm:**

1. **Sort the observed data in ascending order.**
2. **Calculate the theoretical quantiles for the assumed distribution.**
3. **Plot the observed quantiles (y-axis) against the theoretical quantiles (x-axis).**
4. **Draw a reference line (usually y=x).**
5. **Assess how closely the points follow the reference line.**

```python
import scipy.stats as stats

assumed_distribution = stats.expon

# Draw a Q-Q plot for the data
stats.probplot(data, dist=assumed_distribution, plot=plt)
plt.title('Q-Q Plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.grid(True)
plt.show()
```

```python
lambda_ = 1 / np.mean(data)
print('Estimated lambda:', lambda_)
```

```python
ks_test = stats.kstest(data, "expon", args=(0, 1/lambda_))
print('KS test statistic:', ks_test)
```
